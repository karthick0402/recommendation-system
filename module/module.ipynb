```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
x["title"] = le.fit_transform(x["title"])
```

```python
y["genres"] = le.fit_transform(y["genres"])
```

```python
x = x.values
```

```python
y = y.values
```

```python
from sklearn.preprocessing import OneHotEncoder
one = OneHotEncoder()
z = one.fit_transform(y[:,0:1]).toarray()
y = np.delete(y,0,axis = 1)
y = np.concatenate((z,y),axis = 1)
```

```python
y.shape
```

```python
z = one.fit_transform(x[:,0:1]).toarray()
x = np.delete(x,0,axis = 1)
x = np.concatenate((z,x),axis = 1)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state = 0)
```

```python
x_train.shape
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
model = Sequential()
```

```python
model.add(Dense(units = 99,init = 'random_uniform',activation = 'relu'))
```

```python
model.add(Dense(units = 198,init = 'random_uniform',activation = 'relu')) 
```

```python
model.add(Dense(units = 14,init = 'random_uniform',activation = 'softmax')) 
```

```python
model.compile (optimizer = 'adam',loss = 'categorical_crossentropy',metrics = ['accuracy'])
```

```python
model.fit(x_train,y_train,epochs = 5000,batch_size = 128)
```

```python
y_pred = model.predict(x_test)
```

```python
y_pred
```

```python
y_pred = model.predict_classes(x_test)
```
